```python
import pandas as pd
from path import Path
import seaborn as sns
import matplotlib.pyplot as plt
```

# Create data file which combines the clustering results, ratings summary

```python

file_path = Path('cluster_final.csv')
movieclusters_df = pd.read_csv(file_path)
movieclusters_df.drop(columns=['Unnamed: 0'],inplace=True)
movieclusters_df.head()
```

```python
# 
file_path = Path('data_exploration/ratingssummary.csv')
movieratings_df = pd.read_csv(file_path)
movieratings_df.head()
```

```python
visual_df=pd.merge(movieclusters_df,movieratings_df,on='movieid',how='left')
visual_df
```

```python
visual_df.to_csv('visualization_data_final.csv')
```

```python
visual_df.head()
```

```python
# lets get some user tags for our final movieid list
file_path = Path('tags.csv')
tags_df = pd.read_csv(file_path)
tags_df.head()
```

```python
tagsreduced_df=pd.merge(visual_df,tags_df,on='movieid',how='inner')
tagsreduced_df.drop(columns=['Cluster','title','avg_rating','rating_counts','userid','timestamp'],inplace=True)
tagsreduced_df.head()
```
