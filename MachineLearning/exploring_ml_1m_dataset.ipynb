```python
# Initial imports
import pandas as pd
from path import Path
```

```python
file_path = Path('movies.dat')
movies_df = pd.read_csv(file_path,delimiter='::',names=['movieid','title','genres'])
movies_df.head()
```

```python
file_path2 = Path('ratings.dat')
ratings_df = pd.read_csv(file_path2,delimiter='::',names=['userid','movieid','rating','timestamp'])
ratings_df.head()
```

```python
file_path3 = Path('users.dat')
users_df = pd.read_csv(file_path3,delimiter='::',names=['userid','gender','age','occupation','zip-code'])
users_df.head()
```

```python
df=movies_df.merge(ratings_df,on='movieid',how='left')
df.head()
```

```python
df.shape
```

```python
final_df=df.merge(users_df,on='userid',how='left')
final_df.head()
```

```python
final_df.shape
```

```python
# determine how many genre groupings there are and drop any with less than 10 movies.
genrescounts_df=movies_df.groupby('genres').size().reset_index(name='obs').sort_values(['obs'],ascending=False)
genrescounts_df.head(50)
```
