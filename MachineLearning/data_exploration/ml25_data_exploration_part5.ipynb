```python
# Initial imports
import pandas as pd
from path import Path
import numpy as np
import datetime
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# let's take a look at the genome-scores for potentially reducing genome tags
# first import movies
file_path = Path('movies.csv')
movies_df = pd.read_csv(file_path)
movies_df.head()
```

```python
movies_df.shape
```

```python
# We see that there are ~62k movies in this dataframe.  
movies_df.movieId.count()
```

```python
# Now lets import the movie ratings file
file_path2 = Path('ratings.csv')
ratings_df = pd.read_csv(file_path2)
ratings_df.head()
```

```python
# Now let's import the genome-score file which has  ~1k tags with each movie's relevance score to that tag
file_path5 = Path('genome-scores.csv')
genomescores_df = pd.read_csv(file_path5)
genomescores_df.head(5)
```

```python
df=pd.merge(movies_df,genomescores_df,on='movieId',how='left')
```

```python
df.head()
```

```python
X=df.pivot(index='movieId',columns='tagId',values='relevance')
```

```python
X.head()
```

```python
X.shape
```

```python
X.drop(X.columns[0],axis=1,inplace=True)
X.head()
```

```python
X.shape
```

```python
Xcorr=X.corr(method='pearson')
```

```python
Xcorr.to_csv('tagcorrelations.csv')
```

```python
Xcorr_rnd=Xcorr.round()
```

```python
Xcorr_rnd.head()
```

```python
Xcorr_rnd_dr=Xcorr_rnd.loc[(Xcorr_rnd!=0).any(axis=1)]
```

```python
Xcorr_rnd_dr.head()
```

```python
Xtagid_corrsum_df=pd.DataFrame(Xcorr_rnd_dr.sum(),columns=['tagcorelsum'])
```

```python
Xtagid_corrsum_df.sort_values(by='tagcorelsum')
```

```python
Xtagid_goodlist=Xtagid_corrsum_df.loc[Xtagid_corrsum_df['tagcorelsum']>30]
```

```python
Xtagid_goodlist.shape
```

```python
Xtagid_goodlist.to_csv('goodtaglistbycorrel12Feb.csv')
```
